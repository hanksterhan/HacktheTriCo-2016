```python
import pandas as pd
import matplotlib.pyplot as plt

""" Columns:Year,Player,Pos,Age,Tm,G,GS,MP,PER,TS%
            3PAr,FTr,ORB%,DRB%,TRB%,AST%,STL%,BLK%
            TOV%,USG%,blanl,OWS,DWS,WS,WS/48,blank2
            OBPM,DBPM,BPM,VORP,FG,FGA,FG%,3P,3PA,3P%
            2P,2PA,2P%,eFG%,FT,FTA,FT%,ORB,DRB,TRB,AST
            STL,BLK,TOV,PF,PTS
"""

df = pd.read_csv('Seasons_Stats.csv')

kyrie = df.loc[df['Player'] == 'Kyrie Irving']
kyrie = kyrie.set_index([[2012, 2013, 2014, 2015, 2016, 2017]])
del kyrie['Unnamed: 0'], kyrie['Pos'], kyrie['Age'], kyrie['Tm']

kyrie.head()
```

```python
FT = kyrie['FT%']
TS = kyrie['TS%']
years = kyrie['Year']
plt.scatter(FT, TS, cmap=kyrie['Year'])
plt.ticklabel_format(useOffset=False)
i = 0
for x,y in zip(FT,TS):
    year = years.iloc[i]            
    plt.annotate(s = year, xy = (x,y))
    i += 1
# plt.axis([2011,2018,0.84,0.92])
plt.title('Kyrie Irving FT%')
plt.xlabel('FT%')
plt.ylabel('TS%')
plt.margins(0.10)
plt.show()
```
